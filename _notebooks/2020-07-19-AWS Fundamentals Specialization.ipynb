### Going Cloud-Native
    As of the time of publication, the AWS Cloud spans 55 Availability Zones within 18 geographic Regions 
    
    AWS also offers managed compute options, like Amazon Lightsail, that allow you to use compute capacity
        without worrying about provisioning or managing the underlying hardware.
    Amazon Elastic Container Service (Amazon ECS) is a highly scalable, high-performance container 
        orchestration service that supports Docker containers.
    Amazon Elastic Container Service for Kubernetes (Amazon EKS) makes it straightforward to deploy, manage, 
        and scale containerized applications that use Kubernetes on AWS.
    AWS Fargate is a compute engine for Amazon ECS and Amazon EKS that allows you to run containers without
        having to manage servers or clusters.
